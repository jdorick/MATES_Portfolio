# Analysis of 2019 New York City Airbnb Data

```python
import pandas as pd
%matplotlib inline
df = pd.read_csv('./data/AB_NYC_2019.csv', sep=',')
df
```

## Introduction of the Dataset

The dataset used in this project was found from: https://www.kaggle.com/datasets/dgomonov/new-york-city-airbnb-open-data?select=AB_NYC_2019.csv

The dataset used in this project is a collection of Airbnb data of New York City in 2019. The dataset contains the listing and metric data of all Airbnbs in the five boroughs of New York City listed on the official website for the year of 2019. The dataset was chosen because it has a wide variety of data and data that can be divided amongst the five boroughs of New York City to then be compared to one another. On a personal note, the dataset was chosen because I love visiting and exploring New York City as my mother's side of the family was originally from Brooklyn. The dataset contains sixteen columns with the following information in each:

* id: Listing ID on the Airbnb
* name: Name of the listing given by the host
* host_id: ID of the host user
* host_name: Name of the host of the Airbnb
* neighbourhood_group: Which of the five New York City boroughs the Airbnb is located
* neighbourhood: Town of the Airbnb
* latitude: Latitude coordinates
* longitude: Longitude coordinates
* room_type: Space type on the listing
* price: Price per night (in USD)
* minimum_nights: Minimum required nights to purchase the listing
*  number_of_reviews: Number of reviews for the listing
* last_review: Date of most recent (in 2019 or earlier) review
* reviews_per_month: Number of reviews divided by 12
* calculated_host_listings_count: Amount of listings per host
* availability_365: Number of days per year the listing is open

## Analysis of Variable Distributions

### Filtered Data

```python
df_filtered = df.drop(columns=['latitude','longitude'])
df_filtered
```

The filtered dataset does not include the latitude and longitude columns as the values are very close and the neighbourhood and neighbourhood group already give a general estimate of the location of the listing. 

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
df_sort_price = df_filtered[df_filtered['price'] <= 500]
df_sort_price.hist(column='price',alpha=0.90, bins = 30, figsize=(15,10))
plt.ylabel('Number of Airbnbs')
plt.xlabel('Price')
plt.title('Price versus Number of NYC Airbnbs')
```

```python
df_filtered.corr()
```

```python
df2 = df_filtered[['room_type','price']].groupby('room_type', as_index = False).median()
df2.plot.bar(x='room_type',y='price',alpha=0.90,rot=0,figsize=(15,10))
plt.ylabel('Median Price')
plt.xlabel('Room Type')
plt.title('Room Type versus Median Price of NYC Airbnbs')
```

## New Column

```python
# The price per night multiplied by the minimum number of nights required for booking
df_filtered['total_price'] = (df_filtered['price'] * df_filtered['minimum_nights'])
df_filtered
```

## Variable Relationship Analysis

```python
# Total price of every Airbnb against what neighbourhood group it's in
df3 = df_filtered[['neighbourhood_group','total_price']].groupby('neighbourhood_group', as_index = False).median()
df3.plot.bar(x='neighbourhood_group',y='total_price',alpha=0.90,rot=0,figsize=(15,10))
plt.xlabel('Neighbourhood Group')
plt.ylabel('Median Total Price')
plt.title('Median Total Price per NYC Neighbourhood Group')
```

```python
df_boro = df_filtered.filter(items=['neighbourhood_group','total_price'])
df_boro
```

```python
import scipy.stats as stats

# Individual dataframes that will be used as the variables of the ANOVA test for statistical significance
df_brook = df_boro[(df_boro['neighbourhood_group'] == 'Brooklyn')]
df_man = df_boro[(df_boro['neighbourhood_group'] == 'Manhattan')]
df_queen = df_boro[(df_boro['neighbourhood_group'] == 'Queens')]
df_staten = df_boro[(df_boro['neighbourhood_group'] == 'Staten Island')]
df_bronx = df_boro[(df_boro['neighbourhood_group'] == 'Bronx')]

# f-value and p-value of the ANOVA 
fvalue, pvalue = stats.f_oneway(df_brook['total_price'],df_man['total_price'],df_queen['total_price'],df_staten['total_price'],df_bronx['total_price'])

# Print the results
print(fvalue, pvalue)
if pvalue < 0.05:
    print('There is statistical significance in the data between the Airbnb total price and the borough it is in (p < 0.05).')
else:
    print('There is no statistical significance in the data between the Airbnb total price and the borough it is in (p > 0.05).')
```

```python
# Same statistical significance testing but only if the listing of the Airbnb is a private room
private_room = df_filtered[df_filtered['room_type'].str.contains('Private room')]
df_boro2 = private_room.filter(items=['neighbourhood_group','total_price'])
df_boro2
```

```python
# Making the dataframes
df_brook2 = df_boro2[(df_boro2['neighbourhood_group'] == 'Brooklyn')]
df_man2 = df_boro2[(df_boro2['neighbourhood_group'] == 'Manhattan')]
df_queen2 = df_boro2[(df_boro2['neighbourhood_group'] == 'Queens')]
df_staten2 = df_boro2[(df_boro2['neighbourhood_group'] == 'Staten Island')]
df_bronx2 = df_boro2[(df_boro2['neighbourhood_group'] == 'Bronx')]

# Running the ANOVA
fvalue, pvalue = stats.f_oneway(df_brook2['total_price'],df_man2['total_price'],df_queen2['total_price'],df_staten2['total_price'],df_bronx2['total_price'])

print(fvalue, pvalue)
if pvalue < 0.05:
    print('There is statistical significance in the data between the Airbnb total price (for private rooms only) and the borough it is in (p < 0.05).')
else:
    print('There is no statistical significance in the data between the Airbnb total price (for private rooms only) and the borough it is in (p > 0.05).')
```

```python
# Make graphs based on the datasets df_boro and df_boro2
colors = ['olivedrab', 'firebrick', 'steelblue','rebeccapurple','grey']

df_boro.groupby(['neighbourhood_group']).median().plot(kind='pie', y='total_price',figsize=(10,10),colors=colors)
plt.ylabel('')
plt.title('Median Total Prices of all Airbnb Listings Based on its Borough')

df_boro2.groupby(['neighbourhood_group']).median().plot(kind='pie', y='total_price',figsize=(10,10),colors=colors)
plt.ylabel('')
plt.title('Median Total Prices of all Airbnb Private Rooms on its Borough')
```
