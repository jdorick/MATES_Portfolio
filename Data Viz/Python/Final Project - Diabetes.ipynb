```python
# Import Pandas, Seaborn, Numpy, and the Diabetes Dataset
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
df = pd.read_csv('./data/Diabetes.csv', sep=',')
df.head(10)
```

```python
# Check for any null or missing values within the dataset
df.isna().sum()
```

## Convert the Values from Numerical to Groups

```python
# Sex
df['Sex'][df['Sex'] == 0.0] = 'Female'
df['Sex'][df['Sex'] == 1.0] = 'Male'

# Age
df['Age'][df['Age'] == 1.0] = '18 to 24'
df['Age'][df['Age'] == 2.0] = '25 to 29'
df['Age'][df['Age'] == 3.0] = '30 to 34'
df['Age'][df['Age'] == 4.0] = '35 to 39'
df['Age'][df['Age'] == 5.0] = '40 to 44'
df['Age'][df['Age'] == 6.0] = '45 to 49'
df['Age'][df['Age'] == 7.0] = '50 to 54'
df['Age'][df['Age'] == 8.0] = '55 to 59'
df['Age'][df['Age'] == 9.0] = '60 to 64'
df['Age'][df['Age'] == 10.0] = '65 to 69'
df['Age'][df['Age'] == 11.0] = '70 to 74'
df['Age'][df['Age'] == 12.0] = '75 to 79'
df['Age'][df['Age'] == 13.0] = '80 or older'

# Education
df['Education'][df['Education'] == 1.0] = 'No Education'
df['Education'][df['Education'] == 2.0] = 'Elementary'
df['Education'][df['Education'] == 3.0] = 'Some High School'
df['Education'][df['Education'] == 4.0] = 'High School Graduate'
df['Education'][df['Education'] == 5.0] = 'Some College'
df['Education'][df['Education'] == 6.0] = 'College Graduate'

# Income Levels
df['Income'][df['Income'] == 1.0] = '< $10,000'
df['Income'][df['Income'] == 2.0] = '< $15,000'
df['Income'][df['Income'] == 3.0] = '< $20,000'
df['Income'][df['Income'] == 4.0] = '< $25,000'
df['Income'][df['Income'] == 5.0] = '< $35,000'
df['Income'][df['Income'] == 6.0] = '< $50,000'
df['Income'][df['Income'] == 7.0] = '< $75,000'
df['Income'][df['Income'] == 8.0] = '> $75,000'

# Create a separate dataframe that separates the diabetes measures (this is so Tableau works better)
df2 = df
df2['Diabetes_012'][df2['Diabetes_012'] == 0.0] = 'Nondiabetic'
df2['Diabetes_012'][df2['Diabetes_012'] == 1.0] = 'Prediabetic'
df2['Diabetes_012'][df2['Diabetes_012'] == 2.0] = 'Diabetic'
```

```python
df.head(5)
# Convert to a CSV file for later use in Tableau
df.to_csv('Diabetes_Fixed.csv', index=False)
```

```python
# Separate the dataset into three, one for nondiabetic people, prediabetic, and diabetic 
df_nondiabetic = df[df['Diabetes_012'] == 0.0]
df_prediabetic = df[df['Diabetes_012'] == 1.0]
df_diabetic = df[df['Diabetes_012'] == 2.0]
df_diabetic
```

```python
from matplotlib import cm
import matplotlib.pyplot as plt
plt.figure(figsize=(12,10))
# Map color "Blues" because I think it looks the most visually-appealing
sns.heatmap(df.corr(), cmap=cm.Blues, annot=True, center=0)
```

### There is very little correlation amongst a lot of the columns in the dataset

## Demographic Background

```python
# Total of People of Each Age Group
df['Age'].value_counts()
```

```python
# Percentage of nondiabetic individuals with no healthcare
(df_nondiabetic['AnyHealthcare'].count() - df_nondiabetic['AnyHealthcare'].sum()) / (df_nondiabetic['AnyHealthcare'].count()) * 100
```

```python
# Percentage of diabetic individuals with no healthcare
(df_diabetic['AnyHealthcare'].count() - df_diabetic['AnyHealthcare'].sum()) / (df_diabetic['AnyHealthcare'].count()) * 100
```

### Diabetic individuals in the dataset are more likely to have healthcare than nondiabetic individuals

## Health Analysis

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
# Histogram for the BMI between individuals with or without diabetes
df_diabetic.hist(column='BMI', bins = 40, figsize =(15,10))
df_nondiabetic.hist(column='BMI', bins = 40, figsize =(15,10))
```

```python
# Median BMI levels based on Prevalence of Diabetes
print(df_diabetic['BMI'].median())
print(df_prediabetic['BMI'].median())
df_nondiabetic['BMI'].median()
```

### BMI levels steadily increase with the prevalence of diabetes, nondiabetic having the lowest average BMI and diabetic having the highest

```python
# Diabetic Individuals with High Blood Pressure
print(df_diabetic.groupby(['Sex', 'HighBP'])['HighBP'].count())
# Nondiabetic Individuals with High Blood Pressure
df_nondiabetic.groupby(['Sex', 'HighBP'])['HighBP'].count()
```

### Diabetic individuals are far more prone to have high blood pressure than nondiabetic individuals with no difference among the sexes

```python
# Stacked bar chart for the number of diabetic individuals with or without high B.P. based on sex
df_diabetic.groupby(['Sex', 'HighBP']).size().unstack().plot(kind='bar', stacked=True)
# Stacked bar chart for the number of nondiabetic individuals with or without high B.P. based on sex
df_nondiabetic.groupby(['Sex', 'HighBP']).size().unstack().plot(kind='bar', stacked=True)
```

### Stacked bar chart to give the information in the previous groupby visually

```python
# Mean general health score by prevalence of diabetes
avg_health = df[['Diabetes_012','GenHlth']].groupby('Diabetes_012',as_index=False).mean()
avg_health
```
